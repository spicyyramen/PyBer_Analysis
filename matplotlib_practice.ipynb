```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create plot with ax.plt() AKA object-oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# another way to write ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# create quick graph using object-oriented method
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# using plt.show() - create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# create the plot and add label for legend
plt.plot(x_axis,y_axis,label='Boston')

#create lablels for the x and y axes
plt.xlabel("Date")
plt.xlabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0,45)

# create a title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()
```

```python
# create the plot and add label for legend
plt.plot(x_axis,y_axis,marker="*", color="blue", linewidth=2,label='Boston')

#create lablels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the y limit between 0 and 45
plt.ylim(0,45)

# create a title
plt.title("PyBer Fare by Month")

# add a grid
plt.grid()

# add legend
plt.legend()
```

```python
# change line color to green
# change marker to diamond
# change line width to 2 pts
plt.plot(x_axis,y_axis,marker="D", color="green", linewidth=2,label='Boston')

# add legend for city of boston
plt.legend()

# add title and axis labels
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")

# add y-axis limit
plt.ylim(0,45)

# add grid lines
plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create scatter plot using plt.plot
plt.plot(x_axis,y_axis,'o')
```

```python
# plot using scatter method
plt.scatter(x_axis,y_axis)
```

```python
# plot using scatter method
plt.scatter(y_axis,x_axis,c="red",label='Chicago')

# add title and axes labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# limit x-axis data
plt.xlim(0,50)

# invert y-axis to put months in order
plt.gca().invert_yaxis()

# add legend
plt.legend()
```

```python
# bubble chart
plt.scatter(x_axis,y_axis)
```

```python
# multiply each data point in y-axis by 3 
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# use new y-axis to make new plot
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# using list comprehension
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# create scatter plot using o-o method
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# create bubble chart using o-o
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
# use o-o, marker color = skyblue, marker size = data*5, color 20% transparent, black edge circles
# linewidth of circles=2pt
#legend for Boston
fig, ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],c="skyblue",label='Boston',linewidths=2,edgecolors="k",alpha=0.8)

#title
ax.set_title("PyBer Fare by Month")

# axes labels
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

#invert y-axis
ax.invert_yaxis()

# limit x-axis data
ax.set_xlim(0,50)

# add legend
plt.legend()
```

```python
# pie chart using matlab method
plt.subplots(figsize=(8,8))
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode_values,labels=x_axis,autopct='%.1f%%')
```

```python
# assign 12 colors, one for each month
colors = ["slateblue","magenta", "lightblue", "green", "yellowgreen", "greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```

```python
# create piechart using object oriented
fig,ax=plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# format pie chart using object-oriented approach
fig,ax=plt.subplots(figsize=(8,8))

explodes=(0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors=["mediumslateblue","blueviolet","thistle","darkorchid","plum","purple","orchid","mediumvioletred","hotpink","palevioletred","crimson","lightcoral"]
ax.pie(y_axis,
        labels=x_axis,
      explode=explodes,
      shadow=True,
      colors=colors,
      autopct='%.1f%%',
       counterclock=False,
       startangle=95
      )

# add percentage to one decimal place
# increase fig size to 8x8
# explode two hightes percentage months
# add a shadow
# add a start angle so Jan is at top
# reverse order so months run clockwise
# add new colors 
```
